```python
import nltk
```

```python
def gender_features(word):
    return {'last_letter': word[-1]}
```

```python
gender_features('Shrek')
```

```python
from nltk.corpus import names
import random
```

```python
labeled_names = ([(name,'male') for name in names.words('male.txt')]) + ([(name,'female') for name in names.words('female.txt')])
```

```python
random.shuffle(labeled_names)

labeled_names
```

```python
featuresets = [(gender_features(n),gender) for (n,gender) in labeled_names]
train_set, test_set = featuresets[500:], featuresets[:500]
```

```python
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
classifier.classify(gender_features('Neo'))
```

```python
classifier.classify(gender_features('Trinity'))
```

```python
print(nltk.classify.accuracy(classifier, test_set))
```

```python
classifier.show_most_informative_features(10)
```

```python
from nltk.classify import apply_features

train_set = apply_features(gender_features,labeled_names[500:])
test_set = apply_features(gender_features,labeled_names[:500])
```

```python
def gender_features2(name):
    features = {}
    features["first letter"] = name[0].lower()
    features["last letter"] = name[-1].lower()
    for letter in 'abcdefghijklmnopqrstuvwxyz':
        features["count({})".format(letter)] = name.lower().count(letter)
        features["has({})".format(letter)] = (letter in name.lower())
    return features
```

```python
gender_features2('John')
```

```python
featuresets = [(gender_features2(n), gender) for (n,gender) in labeled_names]
train_set, test_set = featuresets[500:],featuresets[:500]
```

```python
classifier = nltk.classify.NaiveBayesClassifier.train(train_set)
```

```python
print(nltk.classify.accuracy(classifier, test_set))
```

```python
train_names = labeled_names[1500:]
devtest_names = labeled_names[500:1500]
test_names = labeled_names[:500]
```

```python
train_set = [(gender_features2(n),gender) for (n,gender) in train_names]
devtest_set = [(gender_features2(n),gender) for (n,gender) in devtest_names]
test_set = [(gender_features2(n),gender) for (n,gender) in test_names]
classifier = nltk.NaiveBayesClassifier.train(train_set)
print(nltk.classify.accuracy(classifier,devtest_set))
```

```python
errors = []
for (name,tag) in devtest_names:
    guess = classifier.classify(gender_features2(name))
    if guess != tag:
        errors.append((tag, guess, name))
```

```python
for (tag, guess, name ) in sorted(errors):
    print('correct={:<8s} guess={:<8s} name={:<30}'.format(tag, guess, name))
```
